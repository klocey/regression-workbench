```python
import pandas as pd
import numpy as np
```

```python
url = 'https://github.com/klocey/hospitals-data-archive/raw/main/dataframes/filtered_files/HAIs/HAIs_'
yr = '2023'

url = url + yr + '.csv'
df = pd.read_csv(url)
df.drop(labels=['Unnamed: 0'], axis=1, inplace=True)
print(df.shape)
df.head()
```

```python
#df.dropna(inplace=True)
nan_rows = df.isnull().sum(axis=1)
nan_cols = df.isnull().sum(axis=0)

#print(nan_rows, 'rows with NaNs')
#print(nan_cols, 'cols with NaNs')
total_nans = df.isnull().sum().sum()
print(total_nans, 'NaNs in the dataframe')
col_with_most_nans = df.isnull().sum().idxmax()
print(col_with_most_nans)
```

```python
mod_df = df.copy(deep=True)
print(mod_df.shape[0], mod_df.shape[1], mod_df.shape[0] * mod_df.shape[1], '\n')

def remove_nans_optimal(df):
    
    total_nans = mod_df.isnull().sum().sum()
    print(total_nans, 'NaNs in df')
    
    while total_nans > 0:
        col_with_most_nans = mod_df.isnull().sum().idxmax()
        print(col_with_most_nans)
        
        #print(mod_df.shape)
        tdf = mod_df.drop(labels=[col_with_most_nans], axis=1)
        #print(tdf.shape)
        tdf.dropna(how='any', axis=0, inplace=True)
        #print(tdf.shape)
        size1 = tdf.shape[0] * tdf.shape[1]
        
        print(tdf.shape[0], tdf.shape[1], size1)
        
        tdf = mod_df.dropna(how='any', axis=0)
        size2 = tdf.shape[0] * tdf.shape[1]
        
        print(tdf.shape[0], tdf.shape[1], size2)
        
        if size1 <= size2:
            mod_df.dropna(how='any', axis=0, inplace=True)
            total_nans = mod_df.isnull().sum().sum()
            
        elif size1 > size2:
            mod_df.drop(labels=[col_with_most_nans], axis=1, inplace=True)
            total_nans = mod_df.isnull().sum().sum()
        
        
    return mod_df

mod_df = remove_nans_optimal(df)
print('\n')
print(mod_df.shape[0], mod_df.shape[1], mod_df.shape[0] * mod_df.shape[1], '\n')

mod_df = df.copy(deep=True)
mod_df.dropna(how='any', axis=0, inplace=True)
print(mod_df.shape[0], mod_df.shape[1], mod_df.shape[0] * mod_df.shape[1], '\n')

mod_df = df.copy(deep=True)
mod_df.dropna(how='any', axis=1, inplace=True)
print(mod_df.shape[0], mod_df.shape[1], mod_df.shape[0] * mod_df.shape[1], '\n')
```
